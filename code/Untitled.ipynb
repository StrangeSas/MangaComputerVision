```python
image_bgr = cv2.imread('../db/buba.png', cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
matrix = np.ones(image_rgb.shape, dtype='uint8') * 50

img_rgb_brighter = cv2.add(image_rgb, matrix)
img_rgb_darker = cv2.subtract(image_rgb, matrix)

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(img_rgb_brighter);plt.title('bright')
plt.subplot(132);plt.imshow(image_rgb);plt.title('norm')
plt.subplot(133);plt.imshow(img_rgb_darker);plt.title('dark')
```

```python
matrix1 = np.ones(image_rgb.shape) * .2
matrix2 = np.ones(image_rgb.shape) * 1.9

img_rgb_brighter = np.uint8(cv2.multiply(np.float64(image_rgb), matrix1))
img_rgb_darker = np.uint8(cv2.multiply(np.float64(image_rgb), matrix2))

plt.figure(figsize=[18,5])
plt.subplot(131);plt.imshow(img_rgb_brighter);plt.title('bright')
plt.subplot(132);plt.imshow(image_rgb);plt.title('norm')
plt.subplot(133);plt.imshow(img_rgb_darker);plt.title('dark')
```
